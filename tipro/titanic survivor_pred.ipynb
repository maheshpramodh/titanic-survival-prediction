```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

en = LabelEncoder()
en.fit(np.array(df['Sex']).reshape(-1,1))
print("Order of appearences for column sex: ",en.classes_)
df['Sex']= en.transform(np.array(df['Sex']).reshape(-1,1))

en.fit(np.array(df['Embarked']).reshape(-1,1))
print("Order of appearences for column Embarked: ",en.classes_)
df['Embarked']= en.transform(np.array(df['Embarked']).reshape(-1,1))
df[['Sex','Embarked']]
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest= train_test_split(df[['Sex','Embarked','Age','Pclass']],df['Survived'],test_size=0.4,random_state=2)
xtrain,ytrain
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(criterion='gini')
classifier.fit(xtrain,ytrain)
tree.plot_tree(classifier,filled=True)
```

```python
ypred = classifier.predict(xtest)
ypred
```

```python
from sklearn.metrics import precision_score,accuracy_score,classification_report

precision = precision_score(ytest,ypred)
accuracy = accuracy_score(ytest,ypred)
classification_rep = classification_report(ytest,ypred)
print("PRECISION SCORE: ",precision)
print("Accuracy score: ",accuracy)
print(classification_rep)
```
